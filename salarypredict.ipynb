```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder,OneHotEncoder
```

```python
data = pd.read_csv("Churn_Modelling.csv")
data
```

```python
data_trim = data.drop(columns=['RowNumber','CustomerId','Surname'])
data_trim
```

```python
label_encoder = LabelEncoder()
data_trim['Gender'] = label_encoder.fit_transform(data_trim['Gender'])
data_trim
```

```python
onehot = OneHotEncoder()
geo = onehot.fit_transform(data_trim[['Geography']])
geo
```

```python
geo_col = geo.toarray()
geo_col
onehot.get_feature_names_out()
```

```python
geo_col_df = pd.DataFrame(geo_col,columns=onehot.get_feature_names_out())
geo_col_df
```

```python
data_trim = pd.concat([data_trim.drop(columns='Geography'),geo_col_df],axis=1)
data_trim
```

```python
X = data_trim.drop(columns='EstimatedSalary')
y = data_trim['EstimatedSalary']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.02,random_state=42)
```

```python
scale = StandardScaler()

X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
import pickle

with open("label_encoder.pkl",'wb') as f:
    pickle.dump(label_encoder,f)

with open("onehot.pkl",'wb') as f:
    pickle.dump(onehot,f)

with open("scaler.pkl",'wb') as f:
    pickle.dump(scale,f)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
```

```python
model = Sequential()

model.add(Dense(64,activation='relu',input_shape =(X_train.shape[1],)))
model.add(Dense(32,activation='relu'))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_absolute_error',metrics=['mae'])
```

```python
import datetime

log_dir = "tb_log/fit/"+datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
```

```python
tb_cb = TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
early_cb = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
history = model.fit(
    X_train,y_train,validation_data = (X_test,y_test),epochs = 100,
    callbacks=[tb_cb,early_cb]
)
```

```python
model.save('salary_pred_mode.h5')
```
